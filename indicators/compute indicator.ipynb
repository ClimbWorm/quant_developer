```python
#往df中添加二次平滑的RSI
def AddRSI_T(df,T):
    '''
    由于RSI14是在index = 14时才有，故rsi2应该在index = 15开始才有意义，rsi7应该在index = 20时才有意义
    '''
    RSI_T = [0]* (T - 1 + 14)
    for i in range((T - 1 + 14),len(df)):
        RSI_T_sub = df[(i - T + 1) : (i + 1)].RSI.mean()
        RSI_T.append(RSI_T_sub)
    df['RSI_{}'.format(T)] = RSI_T
```

```python
def RSI_SMA_shortgodown(df,T):
    '''
        df为包含二次平滑完毕的RSI的表
        T为需要研究的区间长度，只要在T时间内有上穿或者下穿就认为有穿
    '''
        
    RSIShortgodown = []
    for i in range(len(df)):
        RSI_short = df.iloc[i:i + T +1,22].tolist()###这边要注意18是短期的平滑的rsi所在的位置，如果df表格不一样，这里需要修改
        RSI_long = df.iloc[i:i + T +1,23].tolist()
    
        #diff = RSI2 - RSI7#存在由正变负的就ok
        diff = [RSI_short[i] - RSI_long[i] for i in range(len(RSI_short))]
        #print(diff)
        for j in range(len(diff) - 1):
            #print(j)
            before_minus_after = diff[j] - diff[j + 1]
            rsi_short = RSI_short[j]
            rsi_long = RSI_long[j]
            #if (before_minus_after >= diff[j]) & (diff[j] >= 0) & (rsi_short >= 50) & (rsi_short <= 100)& (rsi_long >= 50) & (rsi_long <= 100):给rsi限定范围
            if (before_minus_after >= diff[j]) & (diff[j] >= 0):
                RSIShortgodown_sub = 1
                break
            else:
                RSIShortgodown_sub = 0
                continue
        RSIShortgodown.append(RSIShortgodown_sub)
    #print(len(RSIShortgodown))
    df['RSIShortgodown'] = RSIShortgodown
```

```python
#填充BottomDivergence\n",
Analysis_Data = pd.concat([Analysis_Data, pd.DataFrame(columns=['BottomDivergence'])], sort=False)
def FillBottomDivergence(df):
    BottomDivergence = []
    for i in range(len(df)):
        current_time_index = i
        #print(i)
        #计算previous_extreme_time是哪根bar
        try:
            diff = extreme_bar_num - current_time_index#做差后寻找最后一个为负的 
            previous_extreme_time_index = extreme_bar_num[diff < 0][-1]#获取到上一个极值点的bar_num
        except:
            BottomDivergence.append(0)
            continue

        # 获取当前的RSI和Low
        current_RSI = df.loc[current_time_index].RSI
        current_Low = df.loc[current_time_index].Low


        # 获取上一个极值点到当前的这段区间的RSI和Low
        df_period = df[(df.index >= previous_extreme_time_index) & (df.index < current_time_index)]
        period_extreme_Low = df_period.Low.min()
        period_extreme_Low_index = df_period[df_period.Low == period_extreme_Low].index[0]
        period_extreme_RSI = df.loc[period_extreme_Low_index].RSI
        
        
        
        if (current_Low < period_extreme_Low) & (current_RSI > period_extreme_RSI):
            BottomDivergence_sub = 1
            
        else:
            BottomDivergence_sub = 0
            
        BottomDivergence.append(BottomDivergence_sub)
    df['BottomDivergence'] = BottomDivergence
```

```python
#算Ratio1
#获取了有出现两次的bar，也即这个bar_num为最终未被擦除的\n",
def AddRatio1(df1,df2):

    high_or_low_barnum = df1[abs(df1.length) >1e-6].bar_num.tolist()
    # high_or_low_barnum
    # startNum = high_or_low_barnum[:-1] 
    # endNum = high_or_low_barnum[1:]
    startNum = high_or_low_barnum[:]
    endNum = high_or_low_barnum[1:] + [len(df1)]
    ratio1 = []
    for num in df1.bar_num.tolist():
        if num >= 22:#保证df2中有足够的8个数值可以计算\n", 445
            #print(num)
            flag = 1
            for i,j in zip(startNum,endNum):
                if flag:
                    if (num == i) | (num == j):
                        Idx = df2[df2.bar_num == num].index[0]
                        #print(Idx)
                        last5 = []
                        count = 0
                        while count <= 4:
                            last5.append(abs(df2.loc[Idx - 2*count].length))
                            count += 1
                        last5_mean = np.mean(last5)
                        last5_std = np.std(last5)
                        ratio1_sub = (abs(df2.loc[Idx].length) - last5_mean)/last5_std
                        ratio1.append(ratio1_sub)
                        flag = 0
                    elif (num > i) & (num < j):
                        Idx = df2[df2.bar_num == i].index[0]
                        current_Idx = df1[df1.bar_num == num].index[0]
                        first_last5 = abs(df2.loc[Idx].extremum - df1.loc[current_Idx].extremum)
                        last5 = [first_last5]
                        count = 0
                        while count <= 3:
                            last5.append(abs(df2.loc[Idx - 2*count].length))
                            count += 1
                        last5_mean = np.mean(last5)
                        last5_std = np.std(last5)
                        ratio1_sub = (first_last5 - last5_mean)/last5_std
                        ratio1.append(ratio1_sub)
                        flag = 0
                    else:
                        #print(i)
                        continue
        else:
            ratio1.append(0)
            continue
    df1['ratio1'] = ratio1
    print(len(ratio1))
    print(len(df1.bar_num.tolist()))
```

```python
#算Ratio2
#获取了有出现两次的bar，也即这个bar_num为最终未被擦除的\n",
def AddRatio2(df1,df2):
    high_or_low_barnum = df1[abs(df1.length) >1e-6].bar_num.tolist()
    high_or_low_barnum
    # startNum = high_or_low_barnum[:-1]
    # endNum = high_or_low_barnum[1:]
    startNum = high_or_low_barnum[:]
    endNum = high_or_low_barnum[1:] + [len(df1)]
    ratio2 = []
    for num in df1.bar_num.tolist():
        if num >= 22:#保证df2中有足够的8个数值可以计算\n", 445
            #print(num)
            flag = 1
            for i,j in zip(startNum,endNum):
                if flag:
                    if (num == i) | (num == j):
                        Idx = df2[df2.bar_num == num].index[0]
                        #print(Idx)
                        last9 = []
                        count = 0
                        while count <= 8:
                            last9.append(abs(df2.loc[Idx - count].length))
                            count += 1
                        last9_mean = np.mean(last9)
                        last9_std = np.std(last9)
                        ratio2_sub = (abs(df2.loc[Idx].length) - last9_mean)/last9_std
                        ratio2.append(ratio2_sub)
                        flag = 0
                    elif (num > i) & (num < j):
                        Idx = df2[df2.bar_num == i].index[0]
                        current_Idx = df1[df1.bar_num == num].index[0]
                        first_last9 = abs(df2.loc[Idx].extremum - df1.loc[current_Idx].extremum)
                        last9 = [first_last9]
                        count = 0
                        while count <= 7:
                            last9.append(abs(df2.loc[Idx - count].length))
                            count += 1
                        last9_mean = np.mean(last9)
                        last9_std = np.std(last9)
                        ratio2_sub = (first_last9 - last9_mean)/last9_std
                        ratio2.append(ratio2_sub)
                        flag = 0
                    else:
                        #print(i)
                        continue
        else:
            ratio2.append(0)
            continue
    df1['ratio2'] = ratio2
    print(len(ratio2))
    print(len(df1.bar_num.tolist()))
```
